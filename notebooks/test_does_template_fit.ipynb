```python
from rxntools import utils
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
metacyc_df = pd.read_csv('../data/raw/Esther_Heid_MetaCyc_processed.csv')
```

```python
rxn_idx_list = list(metacyc_df['rxn_idx'])
rxn_strs_list = list(metacyc_df['unmapped'])
rxn_rules_list = list(metacyc_df['rule'])
```

```python
idxs_checked = []
total_rxns_checked = 0
templates_fit = 0
for i in range(0, len(rxn_idx_list)):
    if rxn_idx_list[i] not in idxs_checked:
        try:
            fit = utils.does_template_fit(rxn_strs_list[i], rxn_rules_list[i])
            idxs_checked.append(rxn_idx_list[i])
            total_rxns_checked += 1
            if fit:
                templates_fit += 1
            else:
                pass
        except:
            pass
```
