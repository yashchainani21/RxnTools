```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
reactant1 = Chem.MolFromSmiles('Nc(nc1n([C@@H]2O[C@@H]([C@H]([C@H]2O)O)COP([O-])(OP([O-])(OP([O-])([O-])=O)=O)=O)cnc13)[nH]c3=O')
reactant2 = Chem.MolFromSmiles('Nc1c2ncn([C@@H]3O[C@@H]([C@H]([C@H]3O)O)COP([O-])([O-])=O)c2ncn1')
reactants = tuple((reactant1, reactant2))
```

```python
reaction = AllChem.ReactionFromSmarts('[#6:1]-[#8:2].[#8:3]-[#15:4]>>[#6:1]-[#8:3].[#8:2]-[#15:4]')
```

```python
outcomes = reaction.RunReactants(reactants)
```

```python
outcome_num = 15
products = outcomes[outcome_num]
```
